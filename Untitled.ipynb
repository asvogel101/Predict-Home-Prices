```python


from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
categorical = train.nunique()[train.nunique() <26].keys().tolist()
train_raw = pd.get_dummies(data = train, columns = categorical, drop_first=True)


Y=train_raw['SalePrice']

X=train_raw.drop(['SalePrice', 'Id'], axis=1)

test2 = pd.get_dummies(data = test, columns = categorical, drop_first=True)
test2=test2[test2.columns.intersection(X.columns)]
test2
X = X[X.columns.intersection(test2.columns)]
X


X_train, X_test, y_train, y_test = train_test_split(X, np.log(Y), test_size = 0.15, random_state = 42)

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

```

```python
models = []
#models.append(['RidgeClassifier',RidgeClassifier()])
#models.append(['XGBClassifier',XGBClassifier(use_label_encoder=False,objective='binary:logistic',random_state=0,eval_metric='logloss')])
#models.append(['Logistic Regression',LogisticRegression(random_state=0)])
#models.append(['SVM',SVC(random_state=0)])
#models.append(['KNeigbors',KNeighborsClassifier()])
#models.append(['GaussianNB',GaussianNB()])
#models.append(['BernoulliNB',BernoulliNB()])
#models.append(['DecisionTree',DecisionTreeClassifier(random_state=0)])
#models.append(['RandomForest',RandomForestClassifier(random_state=0)])
#models.append(['AdaBoostClassifier',AdaBoostClassifier()])
#models.append(['MLPClassifier',MLPClassifier(random_state = 42, max_iter=1000)])
#models.append(['ExtraTreesClassifier',ExtraTreesClassifier()])
models.append(['CatBoostRegressor', CatBoostRegressor(eval_metric = 'MSLE', verbose = 0, rsm=0.1)])
#models.append(['GradientBoostingClassifier', GradientBoostingClassifier()])
#models.append(['SGDClassifier',SGDClassifier()])
```

```python
lst_1 = []
for m in range(len(models)):
    lst_2 = []
    model = models[m][1]
    model.fit(X_train,y_train.astype(int))
    y_pred = model.predict(X_test)
    #cm = confusion_matrix(y_test,y_pred)
    accuracies = cross_val_score(estimator= model, X = X_train,y = y_train.astype(int), cv=5)

# k-fOLD Validation
    #roc = roc_auc_score(y_test,y_pred)
    mse = np.sqrt(metrics.mean_squared_log_error(y_test.astype(int), y_pred.astype(int)))
    
    print(models[m][0],':')
    print('MSLE is ', mse)
    #print(cm)
    print('Accuracy Score: ',accuracy_score(y_test.astype(int),y_pred.astype(int)))
    print('')
    print('K-Fold Validation Mean Accuracy: {:.2f} %'.format(accuracies.mean()*100))
    print('')
    #print('ROC AUC Score: {:.2f}'.format(roc))
    print('-'*40)
    print('')
    lst_2.append(models[m][0])
    lst_2.append(accuracy_score(y_test.astype(int),y_pred.astype(int)))
    lst_2.append(accuracies.mean()*100)
    #lst_2.append(roc)
    lst_1.append(lst_2)
```
